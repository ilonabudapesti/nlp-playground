Read the BBC News article: UK's Vicky Pollards 'left behind' http://news.bbc.co.uk/1/hi/education/6173441.stm. The article gives the following statistic about teen language: "the top 20 words used, including yeah, no, but and like, account for around a third of all words." 

How many word types account for a third of all word tokens, for a variety of text sources? What do you conclude about this statistic? Read more about this on LanguageLog, at  http://itre.cis.upenn.edu/~myl/languagelog/archives/003993.html.

```python
texts()
sents()
text1[:100]
```

```python
def topthree(text):
    words = [w for w in text if w.isalpha()]
    outcomes = len(words)
    fd = nltk.FreqDist(words)
    top3 = fd.most_common(3)
    sum = 0
    for w in top3:
        sum += w[1]
    return 100 * sum / outcomes 


topthree(text1)
```

```python
# no of word types that account for third of all word tokens
def third(text):
    words = [w for w in text if w.isalpha()]
    third_of_corpus = round(len(words) / 3)
    #print(third_of_corpus)
    fd = nltk.FreqDist(words)
    top50=fd.most_common(50)
    #print(top50)
    sum = 0
    i = 0
    while sum < third_of_corpus:
        sum += top50[i][1]
        i += 1
    return i
    
print(third(text1), third(text2), third(text3), third(text4), third(text5))
print(third(text6), third(text7), third(text8), third(text9))
```

## 14

Define a function supergloss(s) that takes a synset s as its argument and returns a string consisting of the concatenation of the definition of s, and the definitions of all the hypernyms and hyponyms of s.

```python
def supergloss(s):
    definitions = ''
    definitions = s.definition()
    hypers = '; '.join([w.definition() for w in s.hypernyms()])
    hypos = '; '.join([w.definition() for w in s.hyponyms()])
    definitions += ", " + hypers
    definitions += ", " + hypos
    return definitions

supergloss(wn.synset('female.n.02'))
```

## 17

Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.

```python
from nltk.corpus import stopwords

def top50nonstop(text):
    fd = nltk.FreqDist([w for w in text if w.isalpha() and w.lower() not in stopwords.words('english')])
    return fd.most_common(50)

top50nonstop(brown.words(categories = 'humor'))
```

## 18

Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.

```python
from nltk import bigrams

def top50bigrams(text):
    bis = list(bigrams(text))
    bisnonstop = [b for b in bis 
                  if b[0] not in stopwords.words('english') 
                     and b[1] not in stopwords.words('english')]
    return nltk.FreqDist(bisnonstop).most_common(50)

top50bigrams(brown.words(categories = 'science_fiction'))
```

## 20

Write a function word_freq() that takes a word and the name of a section of the Brown Corpus as arguments, and computes the frequency of the word in that section of the corpus.

```python
def word_freq(word, section):
    fd = nltk.FreqDist(brown.words(fileids=section))
    return fd[word]

word_freq('runaway', 'cg22')
```

## 21

Write a program to guess the number of syllables contained in a text, making use of the CMU Pronouncing Dictionary.

```python
from nltk.corpus import cmudict
dir(cmudict)
```

```python
prondict = nltk.corpus.cmudict.dict()
prondict['fire']
```

```python
def num_syll(text):
    result = 0
    for w in text:
        if w in prondict.keys():
            result += len(prondict[w][0])
    return result

num_syll(brown.words(fileids='cg22'))
```

## 22

Define a function hedge(text) which processes a text and produces a new version with the word 'like' between every third word.

```python
def hedge(text):
    for i in range(0, len(text)):
        if (i + 1) % 3 == 0:
            text[i] += ' like'
    return(' '.join(text))

sent = ['a','b','c', 'and', 'this', 'too']

hedge(sent)
```
